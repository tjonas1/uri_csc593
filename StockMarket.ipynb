```python
import pandas as pd
import pandas_datareader as web
```

```python
#Aktienanalyse
all_data = {ticker: web.get_data_yahoo(ticker)
    for ticker in ['DAI.DE','VNA.DE','PAH3.DE','SHA.DE']}
price = pd.DataFrame({ticker:data ['Adj Close']
                    for ticker, data in all_data.items()})
volume=pd.DataFrame({ticker:data['Volume']
                    for ticker, data in all_data.items()})
```

```python
purchase_list = [['DAI.DE', 54.37], ['PAH3.DE', 49.79], ['VNA.DE', 48.72], ['SHA.DE', 48.72]] 
   
purchase = pd.DataFrame(purchase_list, columns = ['Name', 'purchase_price'])
purchase.set_index('Name',inplace=True)
purchase
```

```python
purchase = pd.DataFrame({'Name':['DAI.DE', 'VNA.DE', 'PAH3.DE', 'SHA.DE'], 
                   'purchase_price':[54.37, 48.72, 49.79, 48.72]}) 
purchase.columns = ['Name', 'purchase_price']
purchase.set_index('Name',inplace=True)
purchase
```
