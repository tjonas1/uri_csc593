```python
# Loading packages Numpy and Pandas
import numpy as np
import pandas as pd
#Load Driving File and keep first column as index
drive = pd.read_csv('../uri_csc593/dataset/drive.csv', index_col=0)
```

```python
drive.head()
```

```python
# Remove unwanted characters from Columns
drive.columns = drive.columns.str.replace("[", "_")
drive.columns = drive.columns.str.replace("]","")
drive.columns = drive.columns.str.replace("/","")
drive.columns = drive.columns.str.replace("%","Pctg")
```

```python
# Deleting rows where GPS signal in the beginning was lost
drive = drive[drive.Longitude_deg != 0]
```

```python
drive.head()
```

```python
# Averages for each column
drive.mean()
```

```python
# Average for just one column
drive[['HV Battery Voltage_V']].mean(axis=0)
```

```python
# Alternative way
drive['HV Battery Voltage_V'].mean()
```

```python
# Create list of grouped items
grouped = drive['Vehicle Speed_kmh'].groupby(drive['Is Driving_bool'])
grouped
```

```python
grouped.mean()
```

```python
# Group items and get mean
drive.groupby(['Is Driving_bool'])[['Vehicle Speed_kmh']].mean()
```

```python
# Add column with speed as m_s
drive['Vehicle Speed_m_s'] = drive['Vehicle Speed_kmh']/3.6
# Add column with Acceleration
drive['Acceleration'] = drive['Vehicle Speed_m_s'] - drive['Vehicle Speed_m_s'].shift()
# Add Column with Jerk
# Needs to be added!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# Add column with distance travelled in that second
drive['Dist_in_s'] = drive['Vehicle Speed_m_s']/1609.344
# Add column with cumulative distance
drive['Cumsum_Dist'] = drive['Dist_in_s'].cumsum()
drive.tail(50)
```
